# Wordcloud Analysis Notebook

This notebook demonstrates wordcloud generation and word frequency analysis for:
1. Paper titles
2. Paper abstracts
3. Combined top words from both sources

## Features:
- Generate wordclouds for titles and abstracts
- Save word frequencies to JSON files
- Extract top N words from both sources without duplicates


```python
# Cell 1: Generate wordcloud for titles and save JSON with most frequent words

# Setup and imports
import sys
import os

# Add the src directory to the path
sys.path.append('../src')

# Import wordcloud functions
from wordcloud_utils import generate_titles_wordcloud

print("✅ Libraries imported successfully")
print("✅ Ready to generate titles wordcloud!")

# Generate wordcloud for titles
print("\n🚀 Generating titles wordcloud...")
generate_titles_wordcloud(
    json_file='../data/wumacat_abstracts.json',
    output_file='../wordclouds/titles_wordcloud.png',
    frequencies_file='../wordclouds/titles_word_frequencies.json'
)

print("\n✅ Titles wordcloud generation completed!")
```

```python
# Cell 2: Generate wordcloud for abstracts and save JSON with most frequent words

# Import wordcloud functions
from wordcloud_utils import generate_abstracts_wordcloud

print("✅ Ready to generate abstracts wordcloud!")

# Generate wordcloud for abstracts
print("\n🚀 Generating abstracts wordcloud...")
generate_abstracts_wordcloud(
    json_file='../data/wumacat_abstracts.json',
    output_file='../wordclouds/abstracts_wordcloud.png',
    frequencies_file='../wordclouds/abstracts_word_frequencies.json'
)

print("\n✅ Abstracts wordcloud generation completed!")
```

```python
# Cell 3: Test the new function to extract top N words from both JSON files

# Import the new function
from wordcloud_utils import extract_top_words_from_json_files

print("✅ Ready to test the new function!")

# Test with n_words=10
print("\n🔍 Testing extract_top_words_from_json_files with n_words=10...")

comprehensive_words = extract_top_words_from_json_files(
    titles_file='../wordclouds/titles_word_frequencies.json',
    abstracts_file='../wordclouds/abstracts_word_frequencies.json',
    n_words=10
)

print(f"\n📊 Results:")
print(f"   Total unique words found: {len(comprehensive_words)}")
print(f"   Words: {comprehensive_words}")

# Display the words in a nice format
print(f"\n🏆 Top {len(comprehensive_words)} unique words from both sources:")
print("=" * 50)
for i, word in enumerate(comprehensive_words, 1):
    print(f"{i:2d}. {word}")

print("\n✅ Function test completed!")
```
